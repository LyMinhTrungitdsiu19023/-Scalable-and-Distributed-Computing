```python
import math
import pandas as pd
from nltk.corpus import stopwords
", ".join(stopwords.words('english'))
STOPW = set(stopwords.words('english'))
def remove_stopwords(text):
    return " ".join([word for word in str(text).split() if word not in STOPW])

def computeTF(wordDict, bagOfWords):
    tfDict = {}
    bagOfWordsCount = len(bagOfWords)
    for word, count in wordDict.items():
        tfDict[word] = count / float(bagOfWordsCount)
    return tfDict
def computeIDF(documents):
    N = len(documents)
    idfDict = dict.fromkeys(documents[0].keys(), 0)
    for document in documents:
        for word, val in document.items():
            if val > 0:
                idfDict[word] += 1
    for word, val in idfDict.items():
        idfDict[word] = math.log(N / float(val))
    return idfDict
def computeTFIDF(tfBagOfWords, idfs):
    tfidf = {}
    for word, val in tfBagOfWords.items():
        tfidf[word] = val * idfs[word]
    return tfidf

doc1 = 'the man went out for a walk'
doc2 = 'the children sat around the fire'

docsw1 = remove_stopwords(doc1)
docsw2 = remove_stopwords(doc2)

bagwords1 = docsw1.split(' ')
bagwords2 = docsw2.split(' ')

uniqueWords = set(bagwords1).union(set(bagwords2))
numwords1 = dict.fromkeys(uniqueWords, 0)
for word in bagwords1:
    numwords1[word] += 1
numwords2 = dict.fromkeys(uniqueWords, 0)
for word in bagwords2:
    numwords2[word] += 1

tf1 = computeTF(numwords1, bagwords1)
tf2 = computeTF(numwords2, bagwords2)
idfs = computeIDF([numwords1, numwords2])
tfidf1 = computeTFIDF(tf1, idfs)
tfidf2 = computeTFIDF(tf2, idfs)
df2 = pd.DataFrame([tfidf1, tfidf2])
df2
```

```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
doc1 = 'the man went out for a walk'
doc2 = 'the kid pushes a chair down to a cat'
vectorizer = TfidfVectorizer()
vectors = vectorizer.fit_transform([doc1, doc2])
feature_names = vectorizer.get_feature_names()
dense = vectors.todense()
denselist = dense.tolist()
df = pd.DataFrame(denselist, columns=feature_names)
df
# The values differ slightly because sklearn uses a smoothed version idf and various other little optimizations. In an example with more text, the score for the word the would be greatly reduced.
```
