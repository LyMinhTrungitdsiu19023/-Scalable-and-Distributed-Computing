```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mlxtend.frequent_patterns import apriori, association_rules


dt = pd.read_csv('./store.csv', header=None)
dt.head(10) # print the first 10 rows
items = dt[0].unique() # print number of unique items 
```

```python
items
```

```python
itemset = set(items)
envals = []
for idx, row in dt.iterrows():
 rowset = set(row)
 labels = {}
 uncomms = list(itemset - rowset)
 commons = list(itemset.intersection(rowset))
 for uc in uncomms:
  labels[uc] = 0
 for cm in commons:
  labels[cm] = 1
  envals.append(labels) 
```

```python
envals
```

```python
ohedt = pd.DataFrame(envals)
fitems = apriori(ohedt, min_support=0.001, use_colnames=True, verbose=1, low_memory=True)
fitems.head(10)
```

```python
rules = association_rules(fitems, metric="confidence", min_threshold=0.6)
rules.head()
```

```python
plt.scatter(rules['support'], rules['confidence'], alpha=0.5)
plt.xlabel('support')
plt.ylabel('confidence')
plt.title('Support vs Confidence')
plt.show()
```
